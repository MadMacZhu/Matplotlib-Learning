```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
```

```python
slices = [60, 40, 30, 20]
labels = ['sixty', 'forty', 'extra1', 'extra2']
colors = ['#008fd5', '#fc4f30', '#efae37', '#6d904f']
```

```python
plt.pie(slices,
        labels = labels,
        colors = colors,
        wedgeprops={'edgecolor': 'black'})

plt.title('My Awesome Pie Chart')
plt.tight_layout()
plt.show()
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]
```

```python
plt.pie(slices,
        labels = labels,
        explode = explode,
        shadow = True,
        startangle = 90,
        autopct = '%1.1f%%',
        wedgeprops={'edgecolor': 'black'})

plt.title('My Awesome Pie Chart')
plt.tight_layout()
plt.show()   
```
